```python
# To check if any rows have Nan Value
df.isna().sum()
```

```python
cols_to_fill_zero=["Bedroom2","Bathroom","Car","Distance","Propertycount"]
```

```python
df[cols_to_fill_zero]=df[cols_to_fill_zero].fillna(0)
```

```python
df.isna().sum()
```

```python
df["Landsize"]=df["Landsize"].fillna(df["Landsize"].mean())
df["BuildingArea"]=df["BuildingArea"].fillna(df["BuildingArea"].mean())
df
```

```python
cols_to_use=['Suburb', 'Rooms', 'Type', 'Price', 'Method', 'SellerG',
       'Distance', 'Bedroom2', 'Bathroom', 'Car',
       'Landsize', 'BuildingArea', 'CouncilArea', 
        'Regionname', 'Propertycount']
```

```python
df=df[cols_to_use]
df.shape
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
# Now to create Dummy Variable for text value
df=pd.get_dummies(df,drop_first=True)
df
```

```python
x=df.drop('Price',axis="columns")
```

```python
y=df["Price"]
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(x,y,test_size=0.3,random_state=3)
```

```python
# Basic Linear Regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression().fit(X_train,y_train)
lr.score(X_test,y_test)
```

```python
# L1 Regularization
from sklearn import linear_model
lasso=linear_model.Lasso(alpha=50,max_iter=200,tol=0.1).fit(X_train,y_train)
```

```python
lasso.score(X_test,y_test)
```

```python
lasso.score(X_train,y_train)
```

```python
# L2 Regularization
from sklearn.linear_model import Ridge
ridge=Ridge(alpha=50,max_iter=200,tol=0.1).fit(X_train,y_train)
```

```python
ridge.score(X_test,y_test)
```

```python
ridge.score(X_train,y_train)
```
