```python
import pandas as pd
```

```python
df=pd.read_csv("titanic.csv")
df
```

```python
import math
mean_age=math.ceil( df["Age"].mean())
mean_age
```

```python
df["Age"]=df["Age"].fillna(mean_age)
df
```

```python
from sklearn.preprocessing import LabelEncoder
le_sex=LabelEncoder()
```

```python
df["encoded_sex"]=le_sex.fit_transform(df["Sex"])
df
```

```python
input=df.drop(["PassengerId","Survived","Name","Sex","SibSp","Parch","Ticket","Cabin","Embarked","Fare"],axis="columns")
input
```

```python
target=df.Survived
target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(input,target,test_size=0.2)
```

```python
X_train
```

```python
len(y_train)
```

```python
len(X_test)
```

```python
from sklearn import tree
model =tree.DecisionTreeClassifier(criterion="entropy")
model.fit(X_train.values,y_train)
```

```python
model.score(X_test,y_test)*100
```

```python
model.predict([[2,27.0,1]])
```

```python
from sklearn.metrics import confusion_matrix
y_predict=model.predict(X_test.values)
```

```python
cm=confusion_matrix(y_test,y_predict)
cm
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(2,2))
sn.heatmap(cm,annot=True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
```

```python
df["Predicted"]=model.predict(input)
df.to_csv("New_Predicted.csv",index=False)
```
