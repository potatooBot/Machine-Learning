```python
df=pd.read_csv("carprices.csv")
df
```

```python
x=df[["Mileage","Age(yrs)"]]
x
```

```python
y=df["Sell Prices"]
y
```

```python
#To make the train size of 80% and test size 20 as 0.2 *100=20% i.e test_size=0.2
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```
