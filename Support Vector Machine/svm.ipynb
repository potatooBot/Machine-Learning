```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
df["target"]=iris.target
df
```

```python
#To check how many datapoints have 1 in it
df[df["target"]==1].count()
```

```python
#To check how many datapoints have in it
df[df["target"]==2].count()
```

```python
#lamba function helps to add another column with the help of other column .In this case We are adding a new column flower_name with the help of the target column.
#X is storing the value from target and adding to the new index of flower_name
df["flower_name"]=df["target"].apply(lambda x: iris["target_names"][x])
df.head()
```

```python
# Now making x and y 
y=df.drop(["sepal length (cm)","sepal width (cm)","petal length (cm)","petal width (cm)","flower_name"],axis="columns")
y
```

```python
x=df.drop(["target","flower_name"],axis="columns")
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
len(X_test)
```

```python
# Now importin SVm
from sklearn.svm import SVC
model=SVC(C=10)
model.fit(X_train.values,y_train)
```

```python
model.score(X_test,y_test)*100
```
