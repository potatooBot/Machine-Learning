```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
```

```python
df=pd.read_csv("carprices.csv")
df
```

```python
plt.scatter(df["Mileage"],df["Sell Price($)"],marker="+",color=r"r")
plt.show()
```

```python
plt.scatter(df["Mileage"],df["Sell Price($)"],marker="+",color="r")
plt.scatter(df["Age(yrs)"],df["Sell Price($)"],marker="+",color="b")
```

```python
df
```

```python
dummies=pd.get_dummies(df["Car Model"])
dummies
```

```python
df
```

```python
merged=pd.concat([df,dummies],axis="columns")
merged
```

```python
final=merged.drop(["Car Model","Mercedez Benz C class"],axis="columns")
final
```

```python
x=final.drop(["Sell Price($)"],axis="columns")
x
```

```python
y=final["Sell Price($)"]
y
```

```python
model=linear_model.LinearRegression()
model.fit(x.values,y)
```

```python
model.predict([[45000,4,0,0]])
```

```python
model.predict([[86000,7,0,1]])
```

```python
model.score(x.values,y)
```

```python
plt.scatter(df["Mileage"],df["Sell Price($)"],marker="+",color="r")
plt.scatter(df["Age(yrs)"],df["Sell Price($)"],marker="+",color="b")
plt.plot(df["Mileage"],model.predict([[45000,4,0,0]]))
plt.plot(df["Age(yrs)"],model.predict([[45000,4,0,0]]))
```
